```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.insert(1, '../')
```

```python
from solvers.Solver import PoissonSolver
```

```python
a, b = 0, 1
```

```python
nx = 100
```

```python
fct_D = lambda x : x
fct_f = lambda x : 4* (np.pi**2) * x * np.sin(2* np.pi * x) - 2* np.pi * np.cos(2* np.pi * x)
```

```python
x = np.linspace(a, b, nx)
```

```python
D = fct_D(x)
F = fct_f(x)
```

```python
params = {'domain': [a, b],
          'nx': nx,
          'D': D,
          'F': F}
```

```python
Solver = PoissonSolver(params)
```

```python
u = Solver.solve
```

```python
u_true = np.sin(2* np.pi* x)

plt.plot(x, u_true, label = 'true')
plt.plot(x, u, '--', label = 'fd')

plt.legend()
plt.show()
```

```python
D2 = np.ones(nx)
F2 = np.ones(nx)
```

```python
VD = np.stack([D, D2])
VF = np.stack([F, F2])
```

```python
U = Solver.Vsolve(vect = 'D', D = VD)
```

```python
u_true = np.sin(2* np.pi* x)

plt.plot(x, u_true, label = 'true')
plt.plot(x, U[1, :], '--', label = 'fd')

plt.legend()
plt.show()
```
